```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
```

Defining neuron inputs

```python
neuron_inputs = np.linspace(5,10,100)
```

Implementing Activation Function

```python
def unit_step(x):
  return np.where(x>=0,1,0)
```

Unit Step Function - Outputs 1 if the input is greater than or equal to 0

```python
def sign_function(x):
  return np.sign(x)
```

Signum - Outputs 1 if the input is positive, -1 if the input is negative and 0 if zero

```python
def linear_function(x):
  a,b = 0.5,2
  return a*x+b
```

Linear Function - Simple Linear Transformation f(x)=a*x+b

```python
def piecewise_linear_function(x):
  return np.piecewise(x,[x<7,(7<=x) & (x<8),x>=8],[lambda x:2*x,lambda x: -2*x+16,lambda x:x-6])
```

Piecewise Linear Function - A function composed of linear segments with different slopes

```python
def logistic_function(x):
  return 1/(1+np.exp(-x))
```

Logistic Function - Squares the input leave between 0 and 1, suitable for binary classifications

```python
def hyperbolic_tangent_function(x):
  return np.tanh(x)
```

Hyperbolic Tangent Function - Input value between -1 and 1, useful for neural networks

Creating a perceptron to demonstrate

a. AND operation
b. OR operation
c. XOR operation

Using two input neurons and one output neuron, W1=0.5, w2=0.243, w0 = 0.2.

```python
W = np.array([0.2, 0.5, 0.243]) # Weights for input neurons and bias
learning_rate = 0.1
epochs = 100
```

```python
def step_function(x):
    return np.where(x >= 0, 1, 0)
```

Function to train the perceptron

```python
def train_perceptron(inputs, target):
    global W
    for epoch in range(epochs):
        error_sum = 0
        for i in range(len(inputs)):
            input_values = np.insert(inputs[i], 0, 1)  # Add bias input
            target_output = target[i]
            predicted_output = step_function(np.dot(W, input_values))
            error = target_output - predicted_output
            W += learning_rate * error * input_values
            error_sum += abs(error)
        if error_sum == 0:
            print(f"Converged at epoch {epoch + 1}")
            break
```

Function to test the trained perceptron

```python
def test_perceptron(inputs):
    results = []
    for i in range(len(inputs)):
        input_values = np.insert(inputs[i], 0, 1)
        result = step_function(np.dot(W, input_values))
        results.append(result)
    return results
```

AND Operation

```python
and_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
and_targets = np.array([0, 0, 0, 1])
train_perceptron(and_inputs, and_targets)
and_results = test_perceptron(and_inputs)
print("AND Operation Results:")
print("Expected: ", and_targets)
print("Predicted:", np.array(and_results).flatten())
print("Error:", abs(and_targets - np.array(and_results).flatten()))
```

**Inference:** The perceptron successfully learned the AND operation. All prediction matches the expected values and the eror is zero for all cases.

OR Operation

```python
or_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
or_targets = np.array([0, 1, 1, 1])
train_perceptron(or_inputs, or_targets)
or_results = test_perceptron(or_inputs)
print("OR Operation Results:")
print("Expected: ", or_targets)
print("Predicted:", np.array(or_results).flatten())
print("Error:", abs(or_targets - np.array(or_results).flatten()))
```

**Inference:** The perceptron successfully learned the OR operation.


XOR Operation

```python
xor_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
xor_targets = np.array([0, 1, 1, 0])
train_perceptron(xor_inputs, xor_targets)
xor_results = test_perceptron(xor_inputs)
print("XOR Operation Results:")
print("Expected: ", xor_targets)
print("Predicted:", np.array(xor_results).flatten())
print("Error:", abs(xor_targets - np.array(xor_results).flatten()))
```

**Inference:** The XOR operation poses a challenge for a single layer perceptron due to its non linearity. Did not converge within 100 epochs and the error remains non zero for some cases.

XOR requires a more complex neural network architecture, such as MLP.
